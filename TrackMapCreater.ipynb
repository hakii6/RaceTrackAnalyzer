```python
# 2m : 100m
# height: 0.02m : 1 -> 10m : 500 -10m : -500 
# 10m : 1

from PIL import Image, ImageDraw
import numpy

import json

def Creater(total_distance, race_track_start, race_track_end):
    with open(race_track_start + '.json') as track_file:
        track_data = json.load(track_file)
    pos = track_data["pos"]
    
    
    pos_tmp = {}
    for i in pos.keys():
        if i == 0:
            continue
#         print(i)
        pos_tmp[str(int(i) + 189430)] = pos[i]
        
    for i in pos_tmp.keys():
        if i == "189430":
            continue
        else:
            pos[i] = pos_tmp[i]
        
    # reverse
    
    pos_reverse = {}
    for i in pos.keys():
        pos_reverse[str(int(i) * -1)] = pos[i]
        
    new_track = {}

    for i in reversed(pos_reverse.keys()):
        if int(i) < -320000:
            continue
        else:
            new_track[int(i) + 320000] = pos_reverse[i]
            
    for j in new_track.keys():
        print(j , new_track[j])
#     print(new_track)
        
    with open('Kyoto_outside_double.json', 'w') as new_track_file:
        json.dump(new_track, new_track_file)
    
total_distance = 3200
race_track_start = "Kyoto_outside"
race_track_end = "Kyoto_outside"
direction = "right"

Creater(total_distance, race_track_start, race_track_end)
```
